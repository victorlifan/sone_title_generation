```python
df4 = pd.read_csv('comb_prediction_100.csv')
df4
```

```python
df4 = df4.fillna('')
df4
```

```python
score_1 = 0
score_2 = 0
score_3 = 0
score_auto = 0
for idx in range(len(df4)):
    row = df4.iloc[idx]
    s1 = fuzz.ratio(row['Actual Title'], row['V1_Generated Title'])
    s2 = fuzz.ratio(row['Actual Title'], row['V2_Generated Title'])
    s3 = fuzz.ratio(row['Actual Title'], row['V3_Generated Title'])
    s_auto = fuzz.ratio(row['Actual Title'], row['Autotrain_Generated_Title'])
    score_1 += s1
    score_2 += s2
    score_3 += s3
    score_auto += s_auto
score_1 = score_1/len(df4)
score_2 = score_2/len(df4)
score_3 = score_3/len(df4)
score_auto = score_auto/len(df4)
print('V1 average fuzzy wuzzy score: ', score_1)
print('V2 average fuzzy wuzzy score: ', score_2)
print('V3 average fuzzy wuzzy score: ', score_3)
print('Autotrain average fuzzy wuzzy score: ', score_auto)
```
